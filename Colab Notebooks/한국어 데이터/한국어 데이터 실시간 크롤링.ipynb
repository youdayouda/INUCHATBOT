```python
from selenium.webdriver.common.by import By
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python

query='장학금'#이것이 질문임
if'학사'in query:
    num=16
    xp1="//*[@id=\"menu1516_obj4849\"]/div[2]/form[2]/div/table/tbody/tr["
    xp2="]/td[2]/a/strong"
elif '학점교류' in query:
    num=17
    xp1="//*[@id=\"menu1517_obj4850\"]/div[2]/form[2]/div/table/tbody/tr["
    xp2="]/td[2]/a/strong"
elif('일반' in query)|('행사'in query)|('모집' in query):
    num=18
    xp1="//*[@id=\"menu1518_obj9193\"]/div[2]/form[2]/div/table/tbody/tr["
    xp2="]/td[3]/a/strong"
elif('장학금' in query):
    num=19
    xp1="//*[@id=\"menu1519_obj4852\"]/div[2]/form[2]/div/table/tbody/tr["
    xp2="]/td[3]/a/strong"
elif('등록금' in query)|('납부' in query):
    num=20
    xp1="//*[@id=\"menu1520_obj4853\"]/div[2]/form[2]/div/table/tbody/tr["
    xp2="]/td[2]/a/strong"
elif('교육시험' in query):
    num=21
    xp1="//*[@id=\"menu1530_obj4855\"]/div[2]/form[2]/div/table/tbody/tr["
    xp2="]/td[2]/a/strong"
elif('봉사' in query):
    num=22
    xp1="//*[@id=\"menu1531_obj4856\"]/div[2]/form[2]/div/table/tbody/tr["
    xp2="]/td[2]/a/strong"

#등록금, 교육시험, 봉사는 strong 태그가 없음 고민중

url="https://www.inu.ac.kr/inu/15"+str(num)+"/subview.do"

driver.get(url)

```

```python
list=driver.find_elements(By.CLASS_NAME,"notice")
for i in range(0,len(list)):
    title=list[i].find_element(By.CLASS_NAME,"td-subject")
    n=i+1
    print(str(n)+"번 : "+title.text)
```
